```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
# looking at the dataset we can assume that there are a bunch of independent features and hence this is a case of
# multiple linear regression with Profit as the dependent feature
```

```python
# checking the dataset to identify if there are any null values
df.isnull()
```

```python
# dividing the dataset into dependent and independent features
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# loking at the dataset, the state is a categorical feature wuth three unique values
# one hot encoding this into two columns with dummy values
```

```python
states = pd.get_dummies(X['State'],drop_first=True)
```

```python
# dropping the states column from the data frame and adding this new column

X = X.drop('State', axis=1)
```

```python
X = pd.concat([X, states], axis=1)
```

```python
# splitting hte datatset into train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state =0)
```

```python
# running linear regression on the test data
from sklearn.linear_model import LinearRegression
lin_reg  = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
# predicting y using the test results
y_pred = lin_reg.predict(X_test)
```

```python
# testing the efficacy of the model
from sklearn.metrics import r2_score
score =r2_score(y_test, y_pred)
```
